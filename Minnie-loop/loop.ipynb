```python
layer = 7

for x in range(1,8):

    space_num = layer - x
    for y in range(1, space_num + 1):
        print(" ", end="")

    star_num = 2 * x - 1
    for y in range(1, star_num + 1):
        print("#", end="")
    print("")
```
